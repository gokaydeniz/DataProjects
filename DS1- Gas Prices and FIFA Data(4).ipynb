```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv("gas_prices.csv")
```

```python
gas
```

```python
plt.figure(figsize= (9,4))
plt.plot(gas.Year, gas.USA, "b--",label ="USA")
plt.plot(gas.Year, gas.Canada,"r.-", label ="Canada")
plt.plot(gas["Year"], gas.Germany, "g.--",label ="Germany")
plt.plot(gas["Year"], gas["South Korea"], label ="South Korea")
plt.plot(gas["Year"], gas["Australia"], label ="Australia")
plt.xlabel("Years")
plt.ylabel("Gas Prices (US Dollars)")
plt.xticks(gas.Year[::2])      #Plotting every 2 years
plt.title("Gas Prices over years", fontsize = 22)
plt.grid()
plt.legend()
plt.savefig("Gas Prices over Years.png", dpi=5000)
plt.show()
```

```python
gas.Year
```

```python
plt.figure(figsize=(10,5))
for i in gas:
    if i != "Year":
        plt.plot(gas.Year,gas[i],"-", label= i, )
plt.xticks(gas.Year[::2])
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
country = ["USA","Canada", "Germany"]
for i in country:
    if i != "Year":
        plt.plot(gas.Year,gas[i],"-", label= i, )
plt.xticks(gas.Year[::2].tolist()+[2011])      #We have added 2011 also
plt.grid()
plt.legend()
plt.show()
```

### 2nd  Fifa Data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.tail(3)
```

```python
plt.hist(fifa["Overall"])
plt.show()
```

```python
bins = np.arange(0,110,10)
plt.hist(fifa["Overall"], color= "#69f542",bins = bins)
plt.ylabel("Skill Level")
plt.xlabel("Number of Players")
plt.title("Distribution of Player Skill in FIFA 2018", fontsize = 16)
plt.show()
```

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"]== "Right"].count()[0]
left, right, left+ right, fifa["Preferred Foot"].count()
```

```python
plt.pie([left,right], labels=["Left","Right"], 
        colors=["#652365","red"], autopct="%.2f")
plt.legend()
plt.title("Preferred Foot of FIFA Players")
plt.show()
```

```python
fifa.Weight.tail(3)
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x) ==str else x for x in fifa.Weight]
light = fifa.loc[fifa.Weight >= 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >=125)& (fifa.Weight <175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >=175)& (fifa.Weight <200)].count()[0]
heavy = fifa.loc[(fifa.Weight >=200)].count()[0]
weights = [light, light_medium, medium_heavy, heavy]
plt.pie(weights, 
        labels= ["light(under 125)", "light Medium (125-175)", "medium heavy (175-200)", "heavy (upper 200)"],
        autopct="%.2f")
plt.title("Weight Distribution of FIFA Players (in lbs)")
plt.legend(loc = (1,0))
plt.show()
```

```python
plt.figure(figsize=(5,6))
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]
boxes= plt.boxplot([barcelona, madrid, revs],
            labels=["Barcelona","Real Madrid","New England Revolution"],
                  patch_artist=True)
plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA Overall Rating")
plt.xlabel("FIFA Soccer Clubs")
for box in boxes["boxes"]:
    #Setting edge color
    box.set(color="#700075")
    #Setting fill color (need to write patch_artist = True)
    box.set(facecolor="gray")
plt.yticks(np.arange(40,100,5))
plt.grid()
plt.show()
```
